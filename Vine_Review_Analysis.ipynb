```python
# Import Dependencies
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df
```

```python
vine_filtered = vine_df.loc[vine_df["total_votes"] >= 20, :]
vine_filtered.dtypes
```

```python
vine_helpful = vine_filtered.loc[lambda vine_filtered: (vine_filtered["helpful_votes"]) / (vine_filtered["total_votes"]) >=0.5]
vine_helpful
```

```python
vine_helpful_paid = vine_helpful.loc[vine_helpful["vine"] == "Y",:]
vine_helpful_paid
```

```python
vine_helpful_unpaid = vine_helpful.loc[vine_helpful["vine"] == "N",:]
vine_helpful_unpaid
```

```python
total_unpaid_reviews = vine_helpful_unpaid.count()
total_unpaid_reviews
```

```python
unpaid_5_star = vine_helpful_unpaid.loc[vine_helpful_unpaid["star_rating"] == 5,:].count()
unpaid_5_star
```

```python
percentage_5_star = (unpaid_5_star / total_unpaid_reviews) * 100
percentage_5_star
```
